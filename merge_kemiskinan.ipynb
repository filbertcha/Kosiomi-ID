```python
import pandas as pd
import glob
import os

# ambil semua file csv dalam folder
files = glob.glob("./Kemiskinan/Kemiskinan*.csv")

data_list = []

for file in files:
    
    # baca csv dan skip 5 baris pertama
    df = pd.read_csv(file, header=None, skiprows=5)
    
    # ambil kolom yang dibutuhkan
    df = df[[0, 7]]
    
    # ubah nama kolom
    df.columns = ["provinsi", "index_kemiskinan"]
    
    # ambil tahun dari nama file
    tahun = ''.join(filter(str.isdigit, os.path.basename(file)))
    df["tahun"] = tahun
    
    data_list.append(df)

# gabungkan semua dataset
merged = pd.concat(data_list, ignore_index=True)

# simpan hasil
merged.to_csv("kemiskinan_2016_2025.csv", index=False)

print("Penggabungan selesai!")
```
